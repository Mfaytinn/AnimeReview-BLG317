```python
# script
import pandas as pd
df_user_details = pd.read_csv('/home/arsun/projects/AnimeReview-BLG317/data/archive/users-details-2023.csv')
df_user_details
```

```python
table_1 = df_user_details[['Mal ID', 'Username', "Gender", "Birthday", "Location", "Joined", "Days Watched"]]

table_1.rename(columns={'Mal ID': 'user_id'}, inplace=True)
table_1['Joined'] = pd.to_datetime(
    table_1['Joined'], errors='coerce', utc=True
).dt.tz_localize(None).dt.strftime('%Y-%m-%d %H:%M:%S')

table_1['Birthday'] = pd.to_datetime(
    table_1['Birthday'], errors='coerce', utc=True
).dt.date

USER_COUNT = 100000 
table_1 = table_1[table_1["user_id"] < USER_COUNT]
print(table_1.columns)

table_1['Birthday'].replace(pd.NaT, 'NULL', inplace=True)
table_1["Location"].replace(pd.NaT, 'NULL', inplace=True)
table_1["Gender"].replace(pd.NaT, 'NULL', inplace=True)

table_1.to_csv('/home/arsun/projects/AnimeReview-BLG317/data/processed/1users.csv', index=False)
del table_1
```

```python
df_anime = pd.read_csv("/home/arsun/projects/AnimeReview-BLG317/data/archive/anime-filtered.csv")

df_anime.head()
df_anime.columns
```

```python
df_producer.reset_index(inplace=True)
df_producer.rename(columns={"index": "producer_id", "Producers": "producer_name"}, inplace=True)
df_producer
```
